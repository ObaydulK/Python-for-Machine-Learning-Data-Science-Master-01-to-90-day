```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a,b)

# small axes
axes2 = fig.add_axes([0.2, 0.2, 0.25, 0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

plt.show()
```
