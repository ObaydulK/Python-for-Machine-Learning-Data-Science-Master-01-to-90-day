```python
import numpy as np
x = np.arange(0,10)
y = 2 *x
```

```python
import matplotlib.pyplot as plt
fig,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[1].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2, figsize=(2,3), dpi=300) 
axes[1][0].plot(x,y)
axes[0][1].plot(x,y)
axes[0][1].set_title('TITLE 1,1')
fig.suptitle('figure Lavel', fontsize=16)
 
fig.subplots_adjust(wspace=1, hspace= 0.5)
```
