```python
import numpy as np
import pandas as pd
```

```python
m = np.linspace(0, 10, 11)
```

```python
m
```

```python
c = 3 * 10**8
```

```python
E = m* c**2
```

```python
E
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(m, E, color='green', lw=2)
plt.title('E= mc<2')
plt.xlabel('m in g')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(m,E)
 
```

```python
plt.plot(m, E, color='green', lw=2)
plt.title('E= mc<2')
plt.xlabel('m in g')
plt.yscale('log')
plt.grid(which="both", axis="y")
plt.show()
```

```python


labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr']
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21 ]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09 ]

 
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(labels,july16_2007, label='July_16_2007')
axes.plot(labels,july16_2020, label='July_16_2020')
axes.legend()
```

```python
fig,axes= plt.subplots(nrows=2, ncols=1, figsize=(12,8))
axes[0].plot(labels, july16_2007, label='july16_2007')
axes[0].set_title("July_16_2007")
axes[1].plot(labels, july16_2020, label= 'july16_2020')
axes[1].set_title("July_16_2020")
```

```python
flg, ax1 = plt.subplots(figsize=(12,8))

ax1.plot(labels, july16_2007, lw=2, color="green")
ax1.set_ylabel("July_16_2007", fontsize=18, color='blue')
for label in ax1.get_yticklabels():
    label.set_color('green')

ax1.plot(labels, july16_2020, lw=2, color="red")
ax1.set_ylabel("July_16_2020", fontsize=18, color='red')
for label in ax1.get_yticklabels():
    label.set_color('red')
```
