```python
#count a leller 01

def count_dog(st):
    count = 0
    for word in st.lower().split():
        if word == 'dog':
            count+=1
    return count
count_dog('This dog runs faster dog then the other dog dude')
```

```python
# Ticketing system speed

def caught_speeding(speed, is_birthday):

    if is_birthday:
        speeding = speed-5
    else:
        speeding = speed

    if speeding > 80:
        return "Big Ticket"
    elif speeding > 60:
        return "Smart ticket"
    else:
        return "no Ticket"
caught_speeding(80, True)
    
```
